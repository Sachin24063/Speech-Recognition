```python
# Signal Analysis
def plot_signal(audio_file):
    features, sr, y = extract_features(audio_file)
    
    # Plot audio signal and MFCC
    plt.figure(figsize=(12, 6))
    plt.subplot(2, 1, 1)
    librosa.display.waveshow(y, sr=sr)
    plt.title('Audio Signal')
    plt.subplot(2, 1, 2)
    librosa.display.specshow(features.T, sr=sr, x_axis='time')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Mel-frequency cepstral coefficients (MFCC)')
    plt.tight_layout()
    plt.show()
    
plot_signal("./Julia_Gillard/415.wav")
```

```python
# Feature Extraction Phase
def extract_features(audio_file):
    y, sr = librosa.load(audio_file)
    mfcc_features = mfcc(y, sr, nfft=1024)  # Adjusting nfft to avoid warning
    return mfcc_features
```

```python
# Vector Quantization (VQ) Phase
def train_codebook(features, codebook_size=64):
    # Perform vector quantization (VQ) to reduce the dimensionality of feature vectors
    codebook = np.array([np.mean(features[i:i+len(features)//codebook_size], axis=0) for i in range(0, len(features), len(features)//codebook_size)])
    return codebook
```

```python
# HMM Phase
def train_hmm(features, codebook, num_states=3):
    # Map the features to the nearest codeword in the codebook
    quantized_features = np.argmin(np.linalg.norm(features[:, np.newaxis] - codebook, axis=2), axis=1)
    model = hmm.GaussianHMM(n_components=num_states)
    model.fit(quantized_features.reshape(-1, 1))  # Reshape for HMM fitting
    return model
```

```python
# Output Phase
def recognize_text(audio_file, hmm_model, codebook):
    features = extract_features(audio_file)
    # Map the features to the nearest codeword in the codebook
    quantized_features = np.argmin(np.linalg.norm(features[:, np.newaxis] - codebook, axis=2), axis=1)
    predicted_labels = hmm_model.predict(quantized_features.reshape(-1, 1))  # Reshape for HMM prediction
    # Perform further processing based on the predicted labels
    # For example, map the labels to phonemes or words
    return predicted_labels
```

```python
# Directory containing audio clips
audio_dir = "./Julia_Gillard/"
```

```python
# Feature extraction phase
all_features = []
for filename in os.listdir(audio_dir):
    if filename.endswith(".wav"):
        audio_file = os.path.join(audio_dir, filename)
        features = extract_features(audio_file)
        all_features.append(features)  # Use append instead of extend
all_features = np.concatenate(all_features, axis=0)  # Concatenate features
```

```python
# Vector quantization (VQ) phase
codebook = train_codebook(all_features)
```

```python
# HMM phase
hmm_model = train_hmm(all_features, codebook)
```

```python
# Output phase
for filename in os.listdir(audio_dir):
    if filename.endswith(".wav"):
        audio_file = os.path.join(audio_dir, filename)
        predicted_labels = recognize_text(audio_file, hmm_model, codebook)
        print("Predicted labels for", filename, ":", predicted_labels)
```
